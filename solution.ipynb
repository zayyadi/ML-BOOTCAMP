```python
island_df = df[df['ocean_proximity'] == 'ISLAND']
selected_columns = island_df[['housing_median_age', 'total_rooms', 'total_bedrooms']]
```

```python
X = selected_columns.to_numpy()
```

```python
XTX = X.T @ X
```

```python
XTX_inverse = np.linalg.inv(XTX)
```

```python
y = np.array([950, 1300, 800, 1000, 1300])
```

```python
w = XTX_inverse @ X.T @ y
```
